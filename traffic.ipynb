```python
from cars import Car, Simulator, Road
import numpy as np
from matplotlib import pyplot as plt
import seaborn
%matplotlib inline
```

```python
cars = [Car() for _ in range(30)]
```

```python
road = Road(1000)
```

```python
sim = Simulator(cars, road, 1)
```

```python
hist = sim.loop(60).flatten() # sim automatically drops first 60 seconds
```

```python
times = [i for _ in range(30) for i in range(60)]
```

```python
plt.scatter(hist, times)
```
